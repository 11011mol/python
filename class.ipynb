```python
# Mutable & Immutable

# Immutable
# int의 경우 새로 할당하면 두 값의 주소가 다름 --> Immutable(값이 변하는게 아니라 새로 생성된 거임)
x = 5
print(id(x))

x = 10
print(id(x))

# Mutable
# list의 경우 내부에 값을 추가해도 리스트의 주소가 같음 --> Mutable(주소가 같음, 기존 데이터에 추가한거임, 새로 생성 X)
list1 = [1,2,3,4,5]
print(list1)
print(id(list1))

list1.append(6)
print(list1)
print(id(list1))
```

```python
# Shallow copy & Deep copy

# Shallow copy (같은 주소를 가리킴 --> 같은 데이터를 가리킴)
a = [1,2,3,4,5]
print(id(a))

b = a
print(id(b))

# Deep copy (다른 주소를 가리킴 --> 데이터의 복제본을 가리킴)
c = [1,2,3,4,5]
print(id(c))

from copy import deepcopy
d = deepcopy(c)
print(id(d))
```
