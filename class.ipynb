```python
# Mutable & Immutable

# Immutable
# int의 경우 새로 할당하면 두 값의 주소가 다름 --> Immutable(값이 변하는게 아니라 새로 생성된 거임)
x = 5
print(id(x))

x = 10
print(id(x))

# Mutable
# list의 경우 내부에 값을 추가해도 리스트의 주소가 같음 --> Mutable(주소가 같음, 기존 데이터에 추가한거임, 새로 생성 X)
list1 = [1,2,3,4,5]
print(list1)
print(id(list1))

list1.append(6)
print(list1)
print(id(list1))
```

```python
# Shallow copy & Deep copy

# Shallow copy (같은 주소를 가리킴 --> 같은 데이터를 가리킴)
a = [1,2,3,4,5]
print(id(a))

b = a
print(id(b))

# Deep copy (다른 주소를 가리킴 --> 데이터의 복제본을 가리킴)
c = [1,2,3,4,5]
print(id(c))

from copy import deepcopy
d = deepcopy(c)
print(id(d))
```

```python
# 리스트 함수 --> 중요중요중요

target = []

# 값 추가
target.append(1)
target.append(2)
target.append(3)
print(target)

# 값 pop
print(target.pop())
print(target)

# 값 정렬
target = [3,2,1,5,2,4,6,6,3]
target.sort()
print(target)
target.sort(reverse = -1)
print(target)

# 뒤집기
target = [3,2,1,5,2,4,6,6,3]
target.reverse()
print(target)

# 인덱스
target = [3,2,1,5,2,4,6,6,3]
value = target.index(3)
print(value)

# 삽입
target.insert(2, 10) 
print(target)

# 제거
target.remove(5)
print(target)

# 확장(리스트 뒤에 리스트 연장)
target.extend([1,2,3,4,5]) # 리스트 내에 있는 요소들을 기존 리스트에 추가함으로써 연장
print(target)
target.append([1,2,3,4,5]) # 괄호 내에 있는 리스트(타입)을 요소로써 추가
print(target)

# 개수 세기
target = [3,2,1,5,2,4,6,6,3]
print(target.count(6))

# 인덱스 + 제거
del(target[0])
print(target)

# 전체 개수(리스트 길이)
print(len(target))

# 전체 합
target = [3,2,1,5,2,4,6,6,3]
print(sum(target))
```

```python
def spam(eggs):
    eggs.append(1)      # 주소: 1234, eggs는 mutable 객체 --> ham은 여기서 [0, 1]로 바뀜
    print(2, id(eggs))  # 주소: 1234
    eggs = [2, 3]       # eggs에 [2, 3]을 새로 할당 --> 주소가 바뀜
    print(3, id(eggs))  # 주소: 5678
    print(eggs)         # [2, 3]
    
ham = [0]         # 주소: 1234
print(1, id(ham)) # 주소: 1234
spam(ham)
print(ham)        # [0, 1]
```

```python
# 주소는 실제 "값"이 저장된 위치
# 같은 주소를 가리킨다 -> 같은 값을 출력한다!

def sum(a, b):
    return a+b

def swap01(a, b):
    # 함수 내에서 새로 '='를 이용해서 변수를 만들면, 이건 함수 내에서만 쓸 수 있음
    a, b = b, a               # 이 함수 내에서의 a, b는 여기 안에서만 사용하는거임 --> 지역변수
    print(a, b, id(a), id(b)) # 20, 10, 4321, 8765 -> 주소가 새로 할당됨 
    
def swap02(list1):
    # 리스트 append(값 추가)나 내부 값을 인덱싱으로 변경 --> 주소가 안 바뀜: 리스트가 mutable이기 때문에
    list1[0], list1[1] = list1[1], list1[0] # 주소가 안 바뀜! -> 값은 바꿔줌
    list1 = [10, 20, 30]                    # 주소가 바뀜!, 새로운 변수가 할당되는거
    print(list1)                            # [10, 20, 30]
    
# 전역변수
a = 10 # 주소: 1234 
b = 20 # 주소: 5678

result = sum(a, b)
print(result)

print(a, b, id(a), id(b)) # 10, 20, 1234, 5678
swap01(a, b)
print(a, b, id(a), id(b)) # 10, 20, 1234, 5678

list1 = [80, 90]
print(list1)              # [80, 90]

swap02(list1)
print(list1)              # [90, 80]
```

```python
say = "hello"   # 주소 1234 -> 전역변수
say2 = "yes"
def sayhello(): 
    global say
    say = "bye" # 주소 1234 -> 전역변수
    say2 = "no"
    
sayhello()
print(say)  # "bye" # 주소: 1234
print(say2) # "yes" 
```
