```python
# Mutable & Immutable

# Immutable
# int의 경우 새로 할당하면 두 값의 주소가 다름 --> Immutable(값이 변하는게 아니라 새로 생성된 거임)
x = 5
print(id(x))

x = 10
print(id(x))

# Mutable
# list의 경우 내부에 값을 추가해도 리스트의 주소가 같음 --> Mutable(주소가 같음, 기존 데이터에 추가한거임, 새로 생성 X)
list1 = [1,2,3,4,5]
print(list1)
print(id(list1))

list1.append(6)
print(list1)
print(id(list1))
```

```python
# Shallow copy & Deep copy

# Shallow copy (같은 주소를 가리킴 --> 같은 데이터를 가리킴)
a = [1,2,3,4,5]
print(id(a))

b = a
print(id(b))

# Deep copy (다른 주소를 가리킴 --> 데이터의 복제본을 가리킴)
c = [1,2,3,4,5]
print(id(c))

from copy import deepcopy
d = deepcopy(c)
print(id(d))
```

```python
example_list = [1, 2, 3, 4, "5", "6", 1.11, [1,2,3]]
print(example_list)

empty_list = []
print(empty_list)

# 인덱싱(Indexing) 중요
print(example_list[0])
print(example_list[-1])
print(example_list[2]) 

# in & not in 리스트 안에 존재하면 True를, 존재하지 않으면 False를 내보내는 문법 --> 중요
# True, False (bool)
print(2 in example_list)
print(10 in example_list)

# for문으로 출력하기
for i in example_list:
    print(i)
```

```python
# 리스트 함수 --> 중요중요중요

target = []

# 값 추가
target.append(1)
target.append(2)
target.append(3)
print(target)

# 값 pop
print(target.pop())
print(target)

# 값 정렬
target = [3,2,1,5,2,4,6,6,3]
target.sort()
print(target)
target.sort(reverse = -1)
print(target)

# 뒤집기
target = [3,2,1,5,2,4,6,6,3]
target.reverse()
print(target)

# 인덱스
target = [3,2,1,5,2,4,6,6,3]
value = target.index(3)
print(value)

# 삽입
target.insert(2, 10) 
print(target)

# 제거
target.remove(5)
print(target)

# 확장(리스트 뒤에 리스트 연장)
target.extend([1,2,3,4,5]) # 리스트 내에 있는 요소들을 기존 리스트에 추가함으로써 연장
print(target)
target.append([1,2,3,4,5]) # 괄호 내에 있는 리스트(타입)을 요소로써 추가
print(target)

# 개수 세기
target = [3,2,1,5,2,4,6,6,3]
print(target.count(6))

# 인덱스 + 제거
del(target[0])
print(target)

# 전체 개수(리스트 길이)
print(len(target))

# 전체 합
target = [3,2,1,5,2,4,6,6,3]
print(sum(target))
```
